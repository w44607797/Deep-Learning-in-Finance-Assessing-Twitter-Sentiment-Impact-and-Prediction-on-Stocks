```python
import matplotlib.pyplot as plt

# Data from results_10/checkpoint-3000/trainer_state.json
data = [
    {
      "epoch": 1.53,
      "learning_rate": 5e-05,
      "loss": 0.5257,
      "step": 500
    },
    {
      "epoch": 3.07,
      "learning_rate": 4.094202898550725e-05,
      "loss": 0.2965,
      "step": 1000
    },
    {
      "epoch": 4.6,
      "learning_rate": 3.188405797101449e-05,
      "loss": 0.1492,
      "step": 1500
    },
    {
      "epoch": 6.13,
      "learning_rate": 2.282608695652174e-05,
      "loss": 0.0709,
      "step": 2000
    },
    {
      "epoch": 7.67,
      "learning_rate": 1.3768115942028985e-05,
      "loss": 0.0312,
      "step": 2500
    },
    {
      "epoch": 9.2,
      "learning_rate": 4.710144927536232e-06,
      "loss": 0.0092,
      "step": 3000
    }
]

# Extract the data for plotting
epochs = [point['epoch'] for point in data]
learning_rates = [point['learning_rate'] for point in data]
losses = [point['loss'] for point in data]

# Create figure and axis objects with subplots()
fig, ax1 = plt.subplots()

# Plot the learning rate
color = 'tab:red'
ax1.set_xlabel('epoch')
ax1.set_ylabel('learning_rate', color=color)
ax1.plot(epochs, learning_rates, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create another axis for the loss
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('loss', color=color)
ax2.plot(epochs, losses, color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Show the plot
plt.title('Learning Rate and Loss over Epochs')
fig.tight_layout()
plt.show()
```
